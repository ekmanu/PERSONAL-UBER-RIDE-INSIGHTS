```python
file_path = r'C:\Users\USER\Documents\PERSONAL UBER INSIGHTS\Uber Data\Rider\trips_data-0.csv'
df= pd.read_csv(file_path)
def get_rows(df, n: int=5):
    return df.head(n)
result = get_rows(df)
print(result)
```

```python
# Converting datetime columns to Nairobi time
def convert_to_nairobi_time(df, column_name):
    df[column_name] = pd.to_datetime(df[column_name], utc=True)
    df[column_name] = df[column_name].dt.tz_convert('Africa/Nairobi').dt.tz_localize(None)

# Applying conversion to all datetime columns
for col in ['request_time', 'begin_trip_time', 'dropoff_time']:
    convert_to_nairobi_time(df, col)

# Creating unique DateTime entries
unique_dates = pd.DataFrame({
    'full_date_time': pd.concat([df['request_time'], df['begin_trip_time'], df['dropoff_time']]).unique()
})

# Extracting date and time components
unique_dates['datetime_id'] = unique_dates.index + 1  # Assuming IDs start from 1
unique_dates['year'] = unique_dates['full_date_time'].dt.year
unique_dates['month'] = unique_dates['full_date_time'].dt.month
unique_dates['day'] = unique_dates['full_date_time'].dt.day
unique_dates['day_of_week'] = unique_dates['full_date_time'].dt.day_name()
unique_dates['time'] = unique_dates['full_date_time'].dt.time

# Dim_DateTime table with unique records
dim_datetime = unique_dates[['datetime_id', 'full_date_time', 'year', 'month', 'day', 'day_of_week', 'time']]


print("Data cleaned and saved successfully!")
```

```python
# Step 1: Extracting unique product types from the DataFrame
unique_product_types = df['product_type'].unique()

# Step 2: Creating a new DataFrame for the product_type dimension table
product_type_dim = pd.DataFrame({
    'product_type_id': range(1, len(unique_product_types) + 1),
    'product_type': unique_product_types
})

print(product_type_dim)
```

```python
#Extracting unique cities from the DataFrame
unique_cities = df['city'].unique()

# Creating a new DataFrame for the city dimension table
city_dim = pd.DataFrame({
    'city_id': range(1, len(unique_cities) + 1),
    'city': unique_cities
})

# Display the city dimension table
print(city_dim)
```

```python
df['begintrip_lat'] = df['begintrip_lat'].round(3)
df['begintrip_lng'] = df['begintrip_lng'].round(3)

# To check the unique combinations of address and rounded lat/lng
begintrip_location = df[[ 'begintrip_lat', 'begintrip_lng', 'begintrip_address']].drop_duplicates()



# Step 3: Create a new DataFrame for the begintrip_location dimension table
begintrip_location_dim = begintrip_location.reset_index(drop=True)

# Step 4: Assign an ID to each unique begintrip location, starting from 1
begintrip_location_dim['begintrip_location_id'] = begintrip_location_dim.index + 1

# Reorder columns to have the ID first
begintrip_location_dim = begintrip_location_dim[['begintrip_location_id', 'begintrip_lat', 'begintrip_lng', 'begintrip_address']]

begintrip_location_dim.head(30)
```

```python
df['dropoff_lat'] = df['dropoff_lat'].round(3)
df['dropoff_lng'] = df['dropoff_lng'].round(3)

# To check the unique combinations of address and rounded lat/lng
dropoff_location = df[[ 'dropoff_lat', 'dropoff_lng', 'dropoff_address']].drop_duplicates()

# Step 3: Create a new DataFrame for the begintrip_location dimension table
dropoff_location_dim = dropoff_location.reset_index(drop=True)

# Step 4: Assign an ID to each unique begintrip location, starting from 1
dropoff_location_dim['dropoff_location_id'] = dropoff_location_dim.index + 1

# Reorder columns to have the ID first
dropoff_location_dim = dropoff_location_dim[['dropoff_location_id', 'dropoff_lat', 'dropoff_lng', 'dropoff_address']]

dropoff_location_dim.head(30)
```
