```python
import pandas as pd
```

```python
file_path = r'C:\Users\USER\Documents\PERSONAL UBER INSIGHTS\Uber Data\Rider\trips_data-0.csv'
df= pd.read_csv(file_path)
def get_rows(df, n: int=5):
    return df.head(n)
result = get_rows(df)
print(result)
```

```python
#Converting the request time column to datetime
df['request_time'] = pd.to_datetime(df['request_time'], utc=True)

#converting the request time to Nairobi time(UTC +3) and localizing it 
df['request_time'] = df['request_time'].dt.tz_convert('Africa/Nairobi')
df['request_time']= df['request_time'].dt.tz_localize(None)

# Extracting date and time components for request_time
df['request_date'] = df['request_time'].dt.date
df['request_time'] = df['request_time'].dt.time


# Extracting additional date components for better analysis
df['request_year'] = pd.to_datetime(df['request_date']).dt.year
df['request_month'] = pd.to_datetime(df['request_date']).dt.month
df['request_day'] = pd.to_datetime(df['request_date']).dt.day

# Extracting day of the week
df['request_day_of_week'] = pd.to_datetime(df['request_date']).dt.day_name() 

#Converting the request time column to datetime
df['begin_trip_time'] = pd.to_datetime(df['begin_trip_time'], utc=True)

#converting the request time to Nairobi time(UTC +3)
df['begin_trip_time'] = df['begin_trip_time'].dt.tz_convert('Africa/Nairobi')
df['begin_trip_time']= df['begin_trip_time'].dt.tz_localize(None)

# Extract date and time components for begin_trip_time
df['begin_trip_date'] = df['begin_trip_time'].dt.date
df['begin_trip_time'] = df['begin_trip_time'].dt.time

# Extracting additional date components for better analysis
df['begin_trip_year'] = pd.to_datetime(df['request_date']).dt.year
df['begin_trip_month'] = pd.to_datetime(df['request_date']).dt.month
df['begin_trip_day'] = pd.to_datetime(df['request_date']).dt.day

# Extracting day of the week
df['begin_trip_day_of_week'] = pd.to_datetime(df['request_date']).dt.day_name() 

#Converting the request time column to datetime
df['dropoff_time'] = pd.to_datetime(df['dropoff_time'], utc=True)

#converting the request time to Nairobi time(UTC +3)
df['dropoff_time'] = df['dropoff_time'].dt.tz_convert('Africa/Nairobi')
df['dropoff_time']= df['dropoff_time'].dt.tz_localize(None)

# Extracting date and time components for dropoff_time
df['dropoff_date'] = df['dropoff_time'].dt.date
df['dropoff_time'] = df['dropoff_time'].dt.time

# Extracting additional date components for better analysis
df['dropoff_year'] = pd.to_datetime(df['request_date']).dt.year
df['dropoff_month'] = pd.to_datetime(df['request_date']).dt.month
df['dropoff_day'] = pd.to_datetime(df['request_date']).dt.day

# Extracting day of the week
df['dropoff_day_of_week'] = pd.to_datetime(df['request_date']).dt.day_name() 

print("Data cleaned and saved successfully!")
```
